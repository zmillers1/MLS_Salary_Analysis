```python
#import 
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.plotly as py
import plotly.graph_objs as go
print (__version__)
init_notebook_mode(connected=True)


#import files and make df
mls_2013 = pd.read_csv('MLS_Player_Salaries_2013.csv')
mls_2013 = pd.DataFrame(mls_2013)

mls_2014 = pd.read_csv('MLS_Player_Salaries_2014.csv')
mls_2014 = pd.DataFrame(mls_2014)

mls_2015 = pd.read_csv('MLS_Player_Salaries_2015.csv')
mls_2015 = pd.DataFrame(mls_2015)

mls_2016 = pd.read_csv('MLS_Player_Salaries_2016.csv')
mls_2016 = pd.DataFrame(mls_2016)

mls_2017 = pd.read_csv('MLS_Player_Salaries_2017.csv')
mls_2017 = pd.DataFrame(mls_2017)


# fix base salary to a float point
fix_base_2013 = mls_2013['Base Salary'].str.replace('$',"")
fix_base_2013 = fix_base_2013.str.replace(',','')
fix_base_2013 = fix_base_2013.astype(float)

fix_base_2014 = mls_2014['Base Salary'].str.replace('$',"")
fix_base_2014 = fix_base_2014.str.replace(',','')
fix_base_2014 = fix_base_2014.astype(float)

fix_base_2015 = mls_2015['Base Salary'].str.replace('$',"")
fix_base_2015 = fix_base_2015.str.replace(',','')
fix_base_2015 = fix_base_2015.astype(float)

fix_base_2016 = mls_2016['Base Salary'].str.replace('$',"")
fix_base_2016 = fix_base_2016.str.replace(',','')
fix_base_2016 = fix_base_2016.astype(float)

fix_base_2017 = mls_2017['Base Salary'].str.replace('$',"")
fix_base_2017 = fix_base_2017.str.replace(',','')
fix_base_2017 = fix_base_2017.astype(float)

# fix guaranteed compensation to a float point
fix_comp_2013 = mls_2013['Guaranteed Compensation'].str.replace('$',"")
fix_comp_2013 = fix_comp_2013.str.replace(',','')
fix_comp_2013 = fix_comp_2013.astype(float)

fix_comp_2014 = mls_2014['Guaranteed Compensation'].str.replace('$',"")
fix_comp_2014 = fix_comp_2014.str.replace(',','')
fix_comp_2014 = fix_comp_2014.astype(float)

fix_comp_2015 = mls_2015['Guaranteed Compensation'].str.replace('$',"")
fix_comp_2015 = fix_comp_2015.str.replace(',','')
fix_comp_2015 = fix_comp_2015.astype(float)

fix_comp_2016 = mls_2016['Guaranteed Compensation'].str.replace('$',"")
fix_comp_2016 = fix_comp_2016.str.replace(',','')
fix_comp_2016 = fix_comp_2016.astype(float)

fix_comp_2017 = mls_2017['Guaranteed Compensation'].str.replace('$',"")
fix_comp_2017 = fix_comp_2017.str.replace(',','')
fix_comp_2017 = fix_comp_2017.astype(float)

# Change Season to Year
season_13 = mls_2013['Season'].astype(str)
season_13 = pd.to_datetime(season_13)
season_13 = season_13.dt.year

season_14 = mls_2014['Season'].astype(str)
season_14 = pd.to_datetime(season_14)
season_14 = season_14.dt.year

season_15 = mls_2015['Season'].astype(str)
season_15 = pd.to_datetime(season_15)
season_15 = season_15.dt.year

season_16 = mls_2016['Season'].astype(str)
season_16 = pd.to_datetime(season_16)
season_16 = season_16.dt.year

season_17 = mls_2017['Season'].astype(str)
season_17 = pd.to_datetime(season_17)
season_17 = season_17.dt.year
```

```python
# make clean tables
mls_2013['Base Salary'] = fix_base_2013
mls_2013['Guaranteed Compensation'] = fix_comp_2013
mls_2013['Season'] = season_13

mls_2014['Base Salary'] = fix_base_2014
mls_2014['Guaranteed Compensation'] = fix_comp_2014
mls_2014['Season'] = season_14

mls_2015['Base Salary'] = fix_base_2015
mls_2015['Guaranteed Compensation'] = fix_comp_2015
mls_2015['Season'] = season_15

mls_2016['Base Salary'] = fix_base_2016
mls_2016['Guaranteed Compensation'] = fix_comp_2016
mls_2016['Season'] = season_16

mls_2017['Base Salary'] = fix_base_2017
mls_2017['Guaranteed Compensation'] = fix_comp_2017
mls_2017['Season'] = season_17

# make time series table
frames = [mls_2017, mls_2016, mls_2015, mls_2014, mls_2013]
mls_salary = pd.concat(frames)
mls_salary
```

##### 

```python
# make plot tables

comp = mls_salary[['Guaranteed Compensation','Season']].groupby(['Season']).median().reset_index()
comp = comp.sort_values(by='Guaranteed Compensation', ascending=True)
comp
```

```python
#plotly chart

data = [go.Scatter(
        x=comp['Season'],
        y=comp['Guaranteed Compensation'])]
py.iplot(data)
```
